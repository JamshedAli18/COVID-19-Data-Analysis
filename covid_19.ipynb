Which country has the highest number of confirmed cases

```python
df.groupby('Country/Region')['Confirmed'].sum().sort_values(ascending = False).head(1)
```

Which country has the highest number of deaths

```python
df.groupby('Country/Region')['Deaths'].sum().sort_values(ascending = False).head(1)
```

```python
top_10_deaths = df.groupby('Country/Region')['Deaths'].sum().sort_values(ascending=False).head(10)

# Reset index to convert the series back to a DataFrame
top_10_deaths = top_10_deaths.reset_index()

# Create a bar plot using seaborn
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

bar_plot = sns.barplot(x='Deaths', y='Country/Region', data=top_10_deaths, palette='viridis')

# Add a title and labels
plt.title('Top 10 Countries by Total Deaths', fontsize=20)
plt.xlabel('Total Deaths', fontsize=14)
plt.ylabel('Country/Region', fontsize=14)

for container in bar_plot.containers:
    bar_plot.bar_label(container, fmt='%.0f', padding=3)

plt.show()
```

Which country has the highest number of recoveries

```python
df.groupby('Country/Region')['Recovered'].sum().sort_values(ascending = False).head(1)
```

```python
top_10_recovered = df.groupby('Country/Region')['Recovered'].sum().sort_values(ascending=False).head(10)

# Reset index to convert the series back to a DataFrame
top_10_recovered = top_10_recovered.reset_index()

# Create a bar plot using seaborn with a different color palette
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

bar_plot = sns.barplot(x='Recovered', y='Country/Region', data=top_10_recovered, palette='muted')

# Add a title and labels
plt.title('Top 10 Countries by Total Recovered Cases', fontsize=20)
plt.xlabel('Total Recovered Cases', fontsize=14)
plt.ylabel('Country/Region', fontsize=14)


for container in bar_plot.containers:
    bar_plot.bar_label(container, fmt='%.0f', padding=3)

plt.show()
```

What is the trend of confirmed cases over time?

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
daily_cases = df.groupby('ObservationDate')['Confirmed'].sum()

plt.figure(figsize=(10, 6))
daily_cases.plot()
plt.title('Trend of Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
plt.show()
```

How many observations are there for each country

```python
df.groupby('Country/Region')['ObservationDate'].count().sort_values(ascending=False).head(5)
```

```python
top_5_countries = df['Country/Region'].value_counts().head(5)

# Reset index to convert the series back to a DataFrame
top_5_countries = top_5_countries.reset_index()

# Rename columns for clarity
top_5_countries.columns = ['Country/Region', 'Count']

# Create a bar plot using seaborn
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

bar_plot = sns.barplot(x='Count', y='Country/Region', data=top_5_countries, palette='viridis')

# Add a title and labels
plt.title('Top 5 Countries by Number of Observations', fontsize=20)
plt.xlabel('Number of Observations', fontsize=14)
plt.ylabel('Country/Region', fontsize=14)

for container in bar_plot.containers:
    bar_plot.bar_label(container, fmt='%.0f', padding=3)

plt.show()
```

a heatmap for the top 10 countries with the highest number of confirmed cases

```python
top_countries_confirmed = df.groupby('Country/Region')['Confirmed'].sum().nlargest(10).index

# Filter the DataFrame for these countries
df_top_countries = df[df['Country/Region'].isin(top_countries_confirmed)]

# Create a pivot table with 'Country/Region' and 'ObservationDate' as indices
# and 'Confirmed' as values
pivot_table = df_top_countries.pivot_table(index='Country/Region', columns='ObservationDate', values='Confirmed', aggfunc='sum')

# Create the heatmap with a different color palette
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='Blues')  # Change the colormap here, e.g., 'Blues', 'Reds', 'Greens', etc.
plt.title('Heatmap of Confirmed Cases for Top 10 Countries')
plt.xlabel('Date')
plt.ylabel('Country')
plt.show()
```

a heatmap for the top 10 countries with the highest number of Death cases

```python
top_countries_confirmed = df.groupby('Country/Region')['Confirmed'].sum().nlargest(10).index

# Filter the DataFrame for these countries
df_top_countries = df[df['Country/Region'].isin(top_countries_confirmed)]

# Create a pivot table with 'Country/Region' and 'ObservationDate' as indices
# and 'Confirmed' as values
pivot_table = df_top_countries.pivot_table(index='Country/Region', columns='ObservationDate', values='Deaths', aggfunc='sum')

# Create the heatmap with a different color palette
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='twilight')  # Change the colormap here, e.g., 'Blues', 'Reds', 'Greens', etc.
plt.title('Heatmap of Confirmed Cases for Top 10 Countries')
plt.xlabel('Date')
plt.ylabel('Country')
plt.show()
```
